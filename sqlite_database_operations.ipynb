#SQLite Database Operations

```python
#loading packages
from sqlalchemy import create_engine
import sqlite3
```

```python
#Creating a temporary + local DB using SQLITE
conn = sqlite3.connect('health.db')
c = conn.cursor()
```

```python
#Creating a new table within the health.db database file
c.execute('''
              CREATE TABLE healthinfo
                      (
                        bloodpressure integer,
                        heartreate integer,
                        height_inches real,
                        name text,
                        address text,
                        contact text
                        );
            ''')
conn.commit()
```

```python
c.execute('''
  SELECT * FROM healthinfo;
''')

print(c.fetchall())
#There's actually nothing in the table yet as we have no inserted data
```

```python
#Inserting the data into the database, health.db

sql_query = '''
INSERT INTO healthinfo (
  'bloodpressure',
  'heartreate',
  'height_inches',
  'name',
  'address',
  'contact'
  )
  values (
    90,
    85,
    57.25,
    'Jane Doe',
    '68 McDonalds Rd',
    'janedoe@gmail.com'
  );
'''
print(sql_query)
```

```python
c.execute(sql_query)
conn.commit()
```

Checking to see if the row has been inserted

```python
sql_query_2 = """

select *
from healthinfo;

"""

c.execute(sql_query_2)
print(c.fetchall())
```

##Automatic Table Creation

```python
# Create new column that goes from wide to long, where we keep columns 0:8 and then melt the rest into a new column called 'insurance_type'

columnNames = list(df)
idVars = columnNames[:8]
valueVars = columnNames[8:]

stonybrook_modified = df.melt(id_vars=idVars, value_vars=valueVars)

stonybrook_modified.columns

stonybrook_modified.rename(columns={'HospCode':'hospital_name',
            'variable':'insurance_type',
            'Code':'code',
            'Description':'code_description',
             'value':'cost_negotiated',
             'Minimum Negotiated Charge':'cost_minimum',
             'Maximum Negotiated Charge': 'cost_maximum'}, inplace=True)

stonybrook_modified['insurance_type'].value_counts()

stonybrook_modified.sample(10)
```

```python
#Implementing the to_sql function from Pandas
stonybrook_modified.to_sql('healthinfo', conn, if_exists='replace')
```

```python
#Testing to see if it works
query = '''
  select *
  from healthinfo
  limit 10;
'''

output = pd.read_sql(query, conn)
output
```

##Diving Deeper

```python
#Creating a new query that looks for the code 92950, which is the CPT code for a cardiac arrest
query = '''
  select *
  from healthinfo
  where code = '92950'
'''

output = pd.read_sql(query, conn)
output
```

This looks at the pricing for the CPT code for cardiac arrest for each insurance type using a pivot table. This pivot table averages the price of the negotiated price per each insurance for  cardiac arrest.

```python
pivot_data = output.pivot_table(index='insurance_type', columns='code', values='cost_negotiated', aggfunc='mean')
pivot_data
```
