I am choosing not to remove any rows with missing values because it does not make sense to do so.

```python
#Removing White Space and Special Characters
import re
def cleaning(df):
    def clean_name(name):
        cleaned_name = re.sub(r'[^a-zA-Z0-9]', '', name)
        return cleaned_name.lower()

    df.columns = [clean_name(col) for col in df.columns]
    return df

df = cleaning(df)
df.sample(5)
```

#SQLite Database Operations

```python
#loading packages
from sqlalchemy import create_engine
import sqlite3
```

```python
#Creating a temporary + local DB using SQLITE
conn = sqlite3.connect('health.db')
c = conn.cursor()
```

```python
#Creating a new table within the health.db database file
c.execute('''
              CREATE TABLE healthinfo
                      (
                        bloodpressure integer,
                        heartreate integer,
                        height_inches real,
                        name text,
                        address text,
                        contact text
                        );
            ''')
conn.commit()
```

```python
c.execute('''
  SELECT * FROM healthinfo;
''')

print(c.fetchall())
#There's actually nothing in the table yet as we have no inserted data
```

```python
#Inserting the data into the database, health.db

sql_query = '''
INSERT INTO healthinfo (
  'bloodpressure',
  'heartreate',
  'height_inches',
  'name',
  'address',
  'contact'
  )
  values (
    90,
    85,
    57.25,
    'Jane Doe',
    '68 McDonalds Rd',
    'janedoe@gmail.com'
  );
'''
print(sql_query)
```

```python
c.execute(sql_query)
conn.commit()
```

Checking to see if the row has been inserted

```python
sql_query_2 = """

select *
from healthinfo;

"""

c.execute(sql_query_2)
print(c.fetchall())
```
